```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r'C:\Users\chira\Downloads\Dataset_Task2.csv')
print("No Errors, Dataset is uploaded, Go ahead")
```

```python
df.isnull().sum()
```

*Age and Fare (are numeric data and) can be filled with Mean/Median/Mode, Cabin has too many null values, so we will drop the column*

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].mode()[0])
df['Age'] = df['Age'].fillna(df['Age'].mean())
print("Both the null values are filled")
```

```python
df.drop(columns=['Cabin'], inplace=True) 
print("The Cabin column is dropped")
```

```python
df.isnull().sum()
```

No null values, dataset is cleaned

```python
gender_counts = df['Sex'].value_counts()
labels = gender_counts.index
sizes = gender_counts.values

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

ax.set_title('Gender Distribution')
ax.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
sns.histplot(df['Age'], kde=True, bins=20, color='skyblue')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
print("Age data is converted") 
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce') #This din give error before, DON'T RUN this 
print("Fare data is converted")
```
